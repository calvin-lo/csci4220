```python
import numpy as np
import scipy as sp
from scipy import signal
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('image.jpg', 0) # Load in grayscale
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
img_x = sp.signal.convolve(img, sobel_x, mode='same')
plt.figure(figsize=(5,5))
plt.imshow(img_x, 'gray')

PQ = np.array([1024,1024,3])
```

# FFT Image

```python
f = np.fft.fft2(img)
print(f.shape)
```

# FFT Kernel

```python
sobel_x_padded = np.zeros(img.shape)

sobel_x_padded[:sobel_x.shape[0], :sobel_x.shape[1]] = sobel_x

fft_filter = np.fft.fft2(sobel_x_padded)

#fft_shift = np.fft.fftshift(fft_filter)
```

# FFT Filter Image

```python
fft_filter_image = np.multiply(f, fft_filter)
plt.imshow(20*np.log(np.abs(fft_filter_image)+1), cmap = 'gray')
```

# Inverse

```python
f_ishift = np.fft.ifftshift(fft_filter_image)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
plt.figure(figsize=(5,5))
plt.imshow(img_back, 'gray')
```
