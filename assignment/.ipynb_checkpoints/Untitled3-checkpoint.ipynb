```python
#Calvin Lo 100514352
import numpy as np
import scipy as sp
from scipy import signal
import cv2
import matplotlib.pyplot as plt
import argparse
import time
%matplotlib inline
```

```python
def makeSfilters():
    NF = 13
    SUP = 49
    F = np.zeros((SUP,SUP,NF))
    
    F[:,:,0]=makefilter(SUP,2,1)
    F[:,:,1]=makefilter(SUP,4,1)
    F[:,:,2]=makefilter(SUP,4,2)
    F[:,:,3]=makefilter(SUP,6,1)
    F[:,:,4]=makefilter(SUP,6,2)
    F[:,:,5]=makefilter(SUP,6,3)
    F[:,:,6]=makefilter(SUP,8,1)
    F[:,:,7]=makefilter(SUP,8,2)
    F[:,:,8]=makefilter(SUP,8,3)
    F[:,:,9]=makefilter(SUP,10,1)
    F[:,:,10]=makefilter(SUP,10,2)
    F[:,:,11]=makefilter(SUP,10,3)
    F[:,:,12]=makefilter(SUP,10,4)

    return F
    
def makefilter(sup, sigma, tau):
    hsup = (sup-1)/2
    x,y = np.meshgrid(np.linspace(-hsup, hsup, hsup*2+1), np.linspace(-hsup, hsup, hsup*2+1))
    r = np.power((x*x+y*y),0.5)
    f = np.cos(r*(np.pi*tau/sigma)) * np.exp(-(r*r)/(2*sigma*sigma))
    f = f - np.mean(f)
    f = f/np.sum(np.abs(f))
    return f
```

```python
def textureAnalysis(file1, file2):
    F = makeSfilters()
    
    img1 = cv2.imread(file1)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        
    img2 = cv2.imread(file2)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

    response1 = []
    response2 = []
    
    for i in range(0,13):
        response1.append(np.abs(signal.convolve(img1, F[:,:,i], mode='valid')))
        response2.append(np.abs(signal.convolve(img2, F[:,:,i], mode='valid')))
        
    means1 = []
    std1 = []
    avg1 = []
    means2 = []
    std2 = []
    avg2 = []
    
    for i in range(0,13):
        m1, s1 = cv2.meanStdDev(response1[i])
        means1.append(m1[0][0])
        std1.append(s1[0][0])
        avg1.append(np.average(response1[i]))
        
        m2, s2 = cv2.meanStdDev(response2[i])
        means2.append(m2[0][0])
        std2.append(s2[0][0])
        avg2.append(np.average(response2[i]))
        
    means1 = np.asarray(means1)
    means2 = np.asarray(means2)
    std1 = np.asarray(std1)
    std2 = np.asarray(std2)
    avg1 = np.asarray(avg1)
    avg2 = np.asarray(avg2)
        
    means_distance = np.sqrt(np.sum((means1-means2)**2))
    std_distance = np.sqrt(np.sum((std1-std2)**2))
    avg_distance = np.sqrt(np.sum((avg1-avg2)**2))
    
    return (means_distance, std_distance, avg_distance)
```

```python
def texture_matching(means_distance, std_distance, avg_distance, threshold):
    if means_distance > threshold:
        return False
    if std_distance > threshold:
        return False
    if avg_distance > threshold:
        return False
    
    return True
```

# Set 1

```python
parser = argparse.ArgumentParser(description='CSCI 4220U Assignment 2.')
parser.add_argument('imgfile1', help='Image file 1')
parser.add_argument('imgfile2', help='Image file 2')
args = parser.parse_args(['stained_0.jpg', 'stained_1.jpg'])

img1 = cv2.imread(args.imgfile1)
img2 = cv2.imread(args.imgfile2)

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1)
plt.subplot(132)
plt.imshow(img2)

means_distance, std_distance, avg_distance = textureAnalysis(args.imgfile1, args.imgfile2) 

print ("Means Distance: ", means_distance)
print ("Std Distance: ", std_distance)
print ("Average Distance: ", avg_distance)

matched = texture_matching(means_distance, std_distance, avg_distance, 10.0)

print matched
```
