```python
#Calvin Lo 100514352
import numpy as np
import scipy as sp
from scipy import signal
import cv2
import matplotlib.pyplot as plt
import argparse
import time
%matplotlib inline
```

```python
def makeLMfilters():
    SUP = 49
    SCALEX = [np.sqrt(2), np.power(np.sqrt(2),2), np.power(np.sqrt(2),3)]
    NORIENT = 6
    NROTINV = 12
    
    NBAR = len(SCALEX) * NORIENT
    NEDGE = len(SCALEX) * NORIENT
    NF = NBAR + NEDGE + NROTINV
    F = np.zeros((SUP, SUP, NF))
    hsup = (SUP)/2
    x, y = np.meshgrid(np.linspace(-hsup, hsup, hsup*2+1), np.linspace(hsup, -hsup, hsup*2+1))
    xx = []
    yy = []
    for i in range(0, SUP):
        for j in range(0, SUP):
            xx.append(x[j][i])
            yy.append(y[j][i])
            
    orgpts = np.transpose(np.transpose([xx, yy]))
    
    count = 0;
    for scale in range(0, len(SCALEX)):
        for orient in range (0,NORIENT):
            angle = np.pi * orient/NORIENT
            c = np.cos(angle)
            s = np.sin(angle)
            rotpts =  np.dot([[c,-s],[s,c]], orgpts)
            rotpts = np.asarray(rotpts)
            F[:,:,count]=makefilter_LM(SCALEX[scale],0,1,rotpts,SUP)
            F[:,:,count+NEDGE]=makefilter_LM(SCALEX[scale],0,2,rotpts,SUP)
            count=count+1
    
    count = NBAR + NEDGE
    SCALES = np.power(np.sqrt(2),[1,2,3,4])
    for i in range(0,len(SCALES)):
        F[:,:,count] = normalise(gauss_kernel((SUP, SUP), SCALES[i]))
        F[:,:,count+1] =normalise(log_kernel(SUP,SCALES[i]))
        F[:,:,count+2] = normalise(log_kernel(SUP,3*SCALES[i]))
        count=count+3;
        
    return F

def makefilter_LM(scale, phasex, phasey, pts, sup):
    gx = gauss1d(3*scale, 0, pts[0,:], phasex)
    gy = gauss1d(scale, 0, pts[1,:], phasey)
    f = normalise(np.reshape(gx*gy, (sup, sup)))
    return f

def gauss1d(sigma, mean, x, ord):
    x = x - mean
    num = x * x
    variance = np.power(sigma,2)
    denom = 2 * variance
    g = np.exp(-num/denom)/np.power((np.pi*denom), 0.5)
    if ord == 1:
        g = -g * (x/variance)
    elif ord == 2:
        g = g * ((num-variance)/np.power(variance,2))
    return g
                 
def normalise(f):
    f = f - np.mean(f[:])
    f = f/np.sum(np.abs(f[:]))
    return f

def gauss_kernel(shape=(3,3),sigma=0.5):
    m,n = [(ss-1.)/2. for ss in shape]
    y,x = np.ogrid[-m:m+1,-n:n+1]
    h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
    sumh = h.sum()
    if sumh != 0:
        h /= sumh
    return h

def log_kernel(p2, sigma):
    rows = p2
    cols = p2
    
    r2 = (rows-1)/2
    c2 = (cols-1)/2

    x, y = np.meshgrid(np.linspace(-c2,c2, c2*2 + 1), np.linspace(-r2,r2, r2*2 + 1))
    radsqrd = np.power(x,2) + np.power(y,2)
    
    f = -1 / (np.pi * np.power(sigma,4))* (1 - radsqrd / 2 * np.power(sigma,2))*np.exp(-radsqrd/2*np.power(sigma,2))
    f = f - np.mean(f)
    
    return f
def log_ker(p2,p3):
    siz   = (p2-1)/2
    std2   = np.power(p3,2)

    x,y = np.meshgrid(np.linspace(-siz,siz, 2 * siz + 1), np.linspace(-siz, siz, 2 * siz +1))
    arg   = -(x*x + y*y)/(2*std2)

    h = np.exp(arg)
    eps = 0.00000000000000001
    h[h<eps*np.max(h)] = 0

    sumh = np.sum(h)
      
    if sumh != 0:
        h  = h/sumh
      
    h1 = h*(x*x + y*y - 2*std2)/np.power(std2,2)
    h = h1 - np.sum(h1)/np.prod(p2)
      
    return h
```

```python
def makeSfilters():
    NF = 13
    SUP = 49
    F = np.zeros((SUP,SUP,NF))
    
    F[:,:,0]=makefilter(SUP,2,1)
    F[:,:,1]=makefilter(SUP,4,1)
    F[:,:,2]=makefilter(SUP,4,2)
    F[:,:,3]=makefilter(SUP,6,1)
    F[:,:,4]=makefilter(SUP,6,2)
    F[:,:,5]=makefilter(SUP,6,3)
    F[:,:,6]=makefilter(SUP,8,1)
    F[:,:,7]=makefilter(SUP,8,2)
    F[:,:,8]=makefilter(SUP,8,3)
    F[:,:,9]=makefilter(SUP,10,1)
    F[:,:,10]=makefilter(SUP,10,2)
    F[:,:,11]=makefilter(SUP,10,3)
    F[:,:,12]=makefilter(SUP,10,4)

    return F
    
def makefilter(sup, sigma, tau):
    hsup = (sup-1)/2
    x,y = np.meshgrid(np.linspace(-hsup, hsup, hsup*2+1), np.linspace(-hsup, hsup, hsup*2+1))
    r = np.power((x*x+y*y),0.5)
    f = np.cos(r*(np.pi*tau/sigma)) * np.exp(-(r*r)/(2*sigma*sigma))
    f = f - np.mean(f)
    f = f/np.sum(np.abs(f))
    return f
```

```python
def textureAnalysis(file1, file2):
    F = makeSfilters()
    
    img1 = cv2.imread(file1)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        
    img2 = cv2.imread(file2)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

    response1 = []
    response2 = []
    
    for i in range(0,13):
        response1.append(np.abs(signal.convolve(img1, F[:,:,i], mode='valid')))
        response2.append(np.abs(signal.convolve(img2, F[:,:,i], mode='valid')))
        
    means1 = []
    std1 = []
    avg1 = []
    means2 = []
    std2 = []
    avg2 = []
    histx1 = []
    histx2 = []
    
    # calculate average, standard deviation, means, histogram
    for i in range(0,13):
        I1 = response1[i].astype('float32')
        m1, s1 = cv2.meanStdDev(I1)
        a1 = np.average(I1)
        hx1 = cv2.calcHist([I1],[0],None,[256],[0,256])
        hx1 = hx1/np.linalg.norm(hx1)
        means1.append(m1[0][0])
        std1.append(s1[0][0])
        avg1.append(a1)
        histx1 = np.append(histx1,hx1)
        
        I2 = response2[i].astype('float32')
        m2, s2 = cv2.meanStdDev(I2)
        a2 = np.average(I2)
        hx2 = cv2.calcHist([I2],[0],None,[256],[0,256])
        hx2 = hx2/np.linalg.norm(hx2)
        means2.append(m2[0][0])
        std2.append(s2[0][0])
        avg2.append(a2)
        histx2 = np.append(histx2,hx2)

    #vector1 = np.asarray([np.average(means1), np.average(std1),np.average(avg1)])
    #vector2 = np.asarray([np.average(means2),np.average(std2),np.average(means2)])
    #vector1 = np.append(vector1, np.average(histx1))
    #vector2 = np.append(vector2, np.average(histx2))
    
    vector1 = np.append(means1, std1)
    vector2 = np.append(means2, std2)
    vector1 = np.append(vector1, avg1)
    vector2 = np.append(vector2, avg2)    
    vector1 = np.append(vector1, histx1)
    vector2 = np.append(vector2, histx2)
    
    # calculate euclidean distance
    distance = np.sqrt(np.sum((vector1 - vector2)**2))   

    return distance
```

```python
def texture_matching(distance, threshold):
    if distance > threshold:
        return False
    return True
```

```python
parser = argparse.ArgumentParser(description='CSCI 4220U Assignment 2.')
parser.add_argument('imgfile1', help='Image file 1')
parser.add_argument('imgfile2', help='Image file 2')
args = parser.parse_args(['stained_0.jpg', 'stained_2.jpg'])

img1 = cv2.imread(args.imgfile1)
img2 = cv2.imread(args.imgfile2)

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1)
plt.subplot(132)
plt.imshow(img2)

distance = textureAnalysis(args.imgfile1, args.imgfile2) 

print ("Distance: ", distance)

matched = texture_matching(distance, 5.0)

print matched
```
